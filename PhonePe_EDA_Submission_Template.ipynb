#### Duplicate Values

````python
# Dataset Duplicate Value Count

import pandas as pd

# Load datasets
agg_ins = pd.read_csv("agg_insurance.csv")
agg_tr = pd.read_csv("agg_transaction.csv")
agg_us = pd.read_csv("agg_user.csv")
map_ins = pd.read_csv("map_insurance.csv")
map_tr = pd.read_csv("map_transaction.csv")
map_us = pd.read_csv("map_user.csv")
top_ins = pd.read_csv("top_insurance.csv")
top_tr = pd.read_csv("top_transaction.csv")
top_us = pd.read_csv("top_user.csv")

datasets = {
    "agg_insurance": agg_ins,
    "agg_transaction": agg_tr,
    "agg_user": agg_us,
    "map_insurance": map_ins,
    "map_transaction": map_tr,
    "map_user": map_us,
    "top_insurance": top_ins,
    "top_transaction": top_tr,
    "top_user": top_us
}

# Check duplicate rows
for name, df in datasets.items():
    print(f"\n Dataset: {name}")
    print("Duplicate Rows:", df.duplicated().sum())
````

#### Missing Values/Null Values

````python
# Missing Values/Null Values Count

import pandas as pd

# Store all datasets in dictionary
datasets = {
    "agg_insurance": agg_ins,
    "agg_transaction": agg_tr,
    "agg_user": agg_us,
    "map_insurance": map_ins,
    "map_transaction": map_tr,
    "map_user": map_us,
    "top_insurance": top_ins,
    "top_transaction": top_tr,
    "top_user": top_us
}

# Check missing values
for name, df in datasets.items():
    print(f"\n🔍 Dataset: {name}")
    print(df.isnull().sum())
````

Total Missing Values (Quick Summary)

````python
for name, df in datasets.items():
    total_missing = df.isnull().sum().sum()
    print(f"{name}: Total Missing Values = {total_missing}")
````

````python
summary = []

for name, df in datasets.items():
    total_cells = df.size
    missing = df.isnull().sum().sum()

    summary.append([name, total_cells, missing])

summary_df = pd.DataFrame(summary, columns=["Dataset", "Total Cells", "Missing Values"])
print(summary_df)
````

Percentage of Missing Values

````python
for name, df in datasets.items():
    print(f"\n📊 {name} - Missing %")
    print((df.isnull().sum() / len(df)) * 100)
````

````python
# Visualizing the missing values
````

Bar Graph for Missing Values (All Datasets)

````python
import matplotlib.pyplot as plt

for name, df in datasets.items():
    missing = df.isnull().sum()

    plt.figure(figsize=(8,4))
    missing.plot(kind='bar')

    plt.title(f"Missing Values - {name}")
    plt.xlabel("Columns")
    plt.ylabel("Missing Count")

    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
````

Single Combined Bar Graph

````python
total_missing = {}

for name, df in datasets.items():
    total_missing[name] = df.isnull().sum().sum()

# Convert to DataFrame
missing_df = pd.DataFrame(list(total_missing.items()), columns=["Dataset", "Missing Values"])

# Plot
plt.figure(figsize=(10,5))
plt.bar(missing_df["Dataset"], missing_df["Missing Values"])

plt.title("Total Missing Values Across Datasets")
plt.xlabel("Dataset")
plt.ylabel("Missing Count")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````
